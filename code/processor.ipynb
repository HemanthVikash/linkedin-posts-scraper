```python
import PySimpleGUI as sg
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim

import yaml
```

```python
geolocator = Nominatim(timeout=10, user_agent="app.test")

def get_country(location_string):
    print(location_string)
    if location_string != np.nan:
        location = geolocator.geocode(location_string, addressdetails=True)
        return location.raw['address']['country'], location.raw['display_name']

    return np.nan, np.nan
```

```python
df = pd.read_csv('../../scraper/results/posts.csv',index_col=0)

sample = df.head(10)
sample['country'], sample['formatted_location'] = zip(*sample['location'].map(get_country))
sample
```

```python
sample.drop(columns=['location'], inplace=True)
sample.rename({
    'formatted_location': 'location'
}, axis=1, inplace=True)
sample.loc[sample.followers == '500+', 'followers'] = np.nan


sample
```

```python
options = {
    'name': None,
    'profile_type': None, # 'Page' | 'Person' | None
    'country': "United States",       # Location Object
}
filtered = sample.copy()

for key, value in options.items():
    if value is not None:
        filtered = filtered[filtered[key] == value]

filtered
```

```python
df = filtered.copy().reset_index(drop=True)
df
```

```python
current_row = 0
df = filtered.copy().reset_index(drop=True)

def update_csv(values):
    global current_row
    approval_status = 'Approved' if values['_APPROVED_'] else 'Not Approved'
    
    # Update DataFrame with approval status
    df.at[current_row, 'approved'] = approval_status
    
    # Save DataFrame back to CSV
    df.to_csv('updated_posts.csv', index=False)
    
    sg.popup('Approval status updated successfully!')

def update_gui(row):
    # Update GUI with data from the selected row
    print(current_row)
    print(len(df))
    window['_NAME_'].update(df.at[row, 'name'])
    window['_SUBHEADING_'].update(df.at[row, 'subheading'])
    window['_COMPANY_'].update(df.at[row, 'company'])






# Event loop to process events and update GUI
while True:

    # Define layout for GUI
    layout = [
        [sg.Text('LinkedIn Post Approval')],
        [sg.Text('Name:', size=(10, 1)), sg.Text('', size=(30, 1), key='_NAME_')],
        [sg.Text('Subheading:', size=(10, 1)), sg.Text('', size=(30, 1), key='_SUBHEADING_')],
        [sg.Text('Company:', size=(10, 1)), sg.Text('', size=(30, 1), key='_COMPANY_')],
        [sg.Text('Approval Status:', size=(15, 1)), sg.Radio('Approved', 'RADIO1', key='_APPROVED_', default=True), sg.Radio('Not Approved', 'RADIO1', key='_NOT_APPROVED_')],
        [sg.Submit('Update'), sg.Button('Previous'), sg.Button('Next'), sg.Button('Exit')]
    ]
    # Create the GUI window
    window = sg.Window('LinkedIn Post Approval', layout)
    
    # Update GUI with initial row data
    update_gui(current_row)
    
    event, values = window.read()
    
    if event == sg.WINDOW_CLOSED or event == 'Exit':
        break
    elif event == 'Update':
        update_csv(values)
    elif event == 'Next':
        current_row += 1
        if current_row >= len(df):
            current_row = len(df) - 1
        update_gui(current_row)
    elif event == 'Previous':
        current_row -= 1
        if current_row < 0:
            current_row = 0
        update_gui(current_row)

window.close()
```
